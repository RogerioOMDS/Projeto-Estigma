```python
import pandas as pd
import numpy as np
import os
```

```python
path = 'data_paragrafo'
num_links = 128

num_pasta = []
lista_textos = []
textos = []
metaforico = []
```

```python
for x in range(2, num_links + 1):
    lista_pasta = os.listdir(path + f'/pasta{x}')
    for y in lista_pasta:
        lista_textos.append(y)
        print('pasta ', x)
        with open(path + '/' + f'/pasta{x}' + '/' + y,'r') as file:
            data = file.read().replace('\n', '')
            textos.append(data)
            if x < 63:
                metaforico.append(0)
            else:
                metaforico.append(1)
```

```python
textos
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

def plot_history(history):
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    x = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(x, acc, 'b', label='Training acc')
    plt.plot(x, val_acc, 'r', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(x, loss, 'b', label='Training loss')
    plt.plot(x, val_loss, 'r', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
```

```python
df1 = pd.DataFrame(textos, columns=['frases'])
df2 = pd.DataFrame(metaforico, columns=['label'])
df = pd.concat([df1, df2], axis=1)
df
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
max_length = 32
embedding_dim = 100
trunc_type='post'
padding_type='post'
oov_tok = "<OOV>"
training_size = 140
```

```python
from keras.backend import clear_session
clear_session()
```

```python
from sklearn.model_selection import train_test_split

samples = df.shape[0]
indices = np.arange(samples)

X_train, X_test, y_train, y_test, idx1, idx2 = train_test_split(df.iloc[:,0], df.iloc[:,1],indices, test_size=0.3)
```

```python
tokenizer = Tokenizer(oov_token=oov_tok)
tokenizer.fit_on_texts(X_train)

word_index = tokenizer.word_index # quais palavras são contadas

training_sequences = tokenizer.texts_to_sequences(X_train)
training_padded = pad_sequences(training_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)

testing_sequences = tokenizer.texts_to_sequences(X_test)
testing_padded = pad_sequences(testing_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)
```

```python
training_padded = np.array(training_padded)
training_labels = np.array(y_train)
testing_padded = np.array(testing_padded)
testing_labels = np.array(y_test)
```

```python
vocab_size = len(word_index)
vocab_size
```

```python
from keras.models import Sequential
from keras import layers
import tensorflow as tf
from tensorflow.keras.layers import LSTM
```

```python
callback = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=3,
    verbose=1,
    restore_best_weights=True,
)
```

```python
lstm_out = 196

model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(vocab_size + 1, embedding_dim, input_length = max_length))
model.add(tf.keras.layers.SpatialDropout1D(0.4))
model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(lstm_out, dropout=0.05, recurrent_dropout=0.2)))
# model.add(tf.keras.layers.Dense(2, activation='softmax'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
optimizer = tf.keras.optimizers.Adam()
model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
```

```python
model.summary()
```

```python
epochs = 30
history = model.fit(training_padded, training_labels, 
                    batch_size = 128, epochs = epochs, 
                    validation_data=(testing_padded,testing_labels), 
                    callbacks = [callback])
```

```python
metrics = pd.DataFrame(history.history)
metrics
```

```python
plot_history(history)
```

```python
loss, accuracy = model.evaluate(testing_padded, testing_labels, verbose=False)
print(loss)
print(accuracy)
```
